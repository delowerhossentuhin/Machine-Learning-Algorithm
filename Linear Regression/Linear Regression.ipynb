```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
housing_dataset=pd.read_csv('housing_dataset.csv')
housing_dataset
```

```python
# Plot The Data Using Matplotlib
```

```python
X1=housing_dataset['Size'].to_numpy()  # First feature (Size)
X2=housing_dataset['No. of Bedrooms'].to_numpy()  # Second feature (No. of Bedrooms)
Y=housing_dataset['Price ($k)'].to_numpy()  # Target variable (Price), Represent h(x)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X1,Y, color="blue",alpha=1)
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($k)")
plt.title("Size vs. Price")
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(X2,Y,color="red",alpha=1)
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price ($k)")
plt.title("Number of Bedrooms VS Price ($k)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Hypothesis was:
$$ h(x)=Î¸_0+Î¸_1X_1+Î¸_2X_2$$

```python
# lets assume:
Î¸0=50 # intercept
Î¸1=0.1 # Coefficient of Feature 1
Î¸2=10 # Coefficient of Feature 2
```

# Implenting the model using Gradient Descent
$$Î¸_j:=Î¸_j-Î±(\frac{2}{n} \sum_{i=1}^{m} h_Î¸ (x^{(i)})-y^{(i)})x_j$$

```python
learning_rate = 0.00000001  # Small learning rate
epochs = 100  # Number of iterations
n = float(len(X1))  # Number of data points
```

$$\text{Performing Gradient Descent}$$


```python
for i in range(epochs):
    y_pred=Î¸0+Î¸1*X1+Î¸2*X2 # Predicted Value
    # Compute gradients (partial derivatives)
    D_Î¸0 = (2/n) * sum(y_pred - Y)
    D_Î¸1 = (2/n) * sum(X1 * (y_pred - Y))
    D_Î¸2 = (2/n) * sum(X2 * (y_pred - Y))
    # Update parameters
    Î¸0 = Î¸1 - learning_rate * D_Î¸0
    Î¸1 = Î¸1 - learning_rate * D_Î¸1
    Î¸2 = Î¸2 - learning_rate * D_Î¸2
    # Print every epochs
    print(f"Epoch {i}: Î¸0 = { Î¸0},  Î¸1 = { Î¸1},  Î¸2 = {Î¸2}")
```

```python
# Plotting the regression line (only for Size vs. Price)
plt.scatter(X1, Y, label="Actual Prices")
plt.plot(X1, Î¸0 + Î¸1 * X1 + Î¸2 * np.mean(X2), color="red", label="Regression Line")
#np.mean(X2)-->This line plots the regression line for predicting house prices based on size, while holding the number of bedrooms constant at its average value.
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($k)")
plt.title("Linear Regression: Size vs. Price")
plt.legend()
plt.show()
```
